# Correlation with heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix between 2 discrete dimensions or event pages. It is imp in feature selection

```python
df.corr() 
# for correlation feature value needs to be float or int
```

```python
sns.heatmap(df.corr())
```

# JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D Density plot

```python
# Univariate analysis

sns.jointplot(x='tip', y= 'total_bill', data=df, kind='hex')
# finding analysis b/w total_bill and tip
```

```python
sns.jointplot(x='tip', y= 'total_bill', data=df, kind='reg')
# regression line - best fit line
```

# Pairplot

A "pair plot" is also known as scatter plot, in which one variable in the same data row is matched with another variable's value.

Pair plots are just elaborations on this, showing all variables paired with all the other variables. They are similar to scatter plots

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex') #pairplot based on category of feature/field
```

```python
df['smoker'].value_counts()
```

# Dist plot

Dist plot help us to check the distibution of columns feature

```python
sns.distplot(df['tip'])
```

# Categorical plot in seaborn

Seaborn helps us in doing analysis of categorical data plots.
* boxplot
* violinplot
* countplot
* bar plot

```python
# COUNT PLOT
sns.countplot(x='time',data=df)
```

```python
sns.countplot(y='smoker', data=df)
```

```python
#barplot 

sns.barplot(x='total_bill', y='sex', data=df)
```

# Boxplot


A box and whisker plot is a graph that replresents information from a five-number summary. (0, 25, 50, 75 and 100 percentile)

```python
sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

# Violin plot

Violin plot helps in distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
