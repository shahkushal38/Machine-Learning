# Univariate Bivariate and Multivariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read iris dataset
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape #prints the size of dataset
```

# Univariate Analysis

```python
df_setosa =  df.loc[df['species'] == 'setosa'] # get all frames with species as setosa
```

```python
print(df_setosa)
```

```python
df_virginica =  df.loc[df['species'] == 'virginica']
df_versicolor =  df.loc[df['species'] == 'versicolor']
```

```python
# this plot shows sepal_length graphically with a single variable for univariate analysis
# 'o' is to increase the size of dots
# np.zeros_like() helps in plot value with x axis

plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o') 
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']),'o') 
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']),'o') 
plt.xlabel('Sepal Length')
plt.show()
```

# Bivariate Analysis

```python
# for bivariate we can use seaborn to do analysis with two features

sns.FacetGrid(df, hue="species", height=5).map(plt.scatter,"sepal_length","sepal_width").add_legend()

plt.show()
```

# Multivariate Analysis

```python
# simply use seaborn pairplot function
sns.pairplot(df, hue="species", height= 5)
```
